```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("Sanath369/Telugu_sentiment_movie")

model = AutoModelForSequenceClassification.from_pretrained("Sanath369/Telugu_sentiment_movie")
```

```python
from transformers import pipeline

classifier = pipeline(model = "Sanath369/Telugu_sentiment_movie",return_all_scores=True)
data = ["The movie is good and excelllent","It is wonderful to watch this movie"]
result = classifier(data)
## [{'label': 'ENTAILMENT', 'score': 0.98}]
```

```python
result
```

```python
neg = ["ఇంటర్వెల్ ముందు వరకు కథలో డ్రామా & సాగదీత ఎక్కువ. ఇంటర్వెల్ తర్వాత ట్విస్టులు ఎక్కువ. 'పొలిమేర 3' కోసం బలమైన కథను సిద్ధం చేసుకునే క్రమంలో 'పొలిమేర 2' కథలో సస్పెన్స్, థ్రిల్లర్ అంశాలను ఒక స్థాయికి పరిమితం చేశారు. 'మా ఊరి పొలిమేర' అభిమానులు సైతం ఎటువంటి అంచనాలు లేకుండా వెళితే శాటిస్‌ఫై అవుతారు."]
result1 = classifier(neg)
```

```python
result1
```

```python
def rating(neu,pos,neg):
  if(neu>0.7 and pos>1):
    return 4.5
  elif(pos>0.25 and neu<0.5):
    return 4.5
  elif(neu>0.6 and pos>0.25):
    return 4.5
  elif(neu>0.6 and pos>0.2):
    return 4
  elif(neu>0.5 and pos>0.2):
    return 3.5
  elif(neg>0.25 and pos<0.15):
    return 2
  elif(neg>0.2 and pos<0.2):
    return 3
  else:
    return 2.5
```

```python
rate = rating(result1[0][1]['score'],result1[0][2]['score'],result1[0][0]['score'])
print(rate)
```
